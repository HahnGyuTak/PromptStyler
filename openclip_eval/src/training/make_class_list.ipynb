```python
import os

def list_subfolders(base_folder):
    # 하위 폴더들의 목록을 가져옴
    subfolders = [f.name for f in os.scandir(base_folder) if f.is_dir()]
    subfolders = sorted(subfolders)
    # 각 하위 폴더의 이름을 "class1", "class2" 형식으로 출력
    for index, folder_name in enumerate(subfolders, start=1):
        print(f"'{folder_name.lower()}',",end="")
        if index % 10 ==0:
            print()

# 사용 예시: "path/to/base_folder"를 실제 경로로 바꾸세요.
base_folder = '/data/pacs/photo'
list_subfolders(base_folder)
```

```python
from torchvision.datasets import ImageFolder
from torchvision import transforms

# 이미지 폴더의 경로를 설정합니다.
data_dir = '/data/pacs/photo'

# ImageFolder 데이터셋을 만듭니다.
dataset = ImageFolder(root=data_dir, transform=transforms.ToTensor())

# 클래스 폴더들의 순서를 출력합니다.
for index, (file_path, class_idx) in enumerate(dataset.samples):
    print(f"{index+1}: {file_path} - Class {class_idx}")
```
